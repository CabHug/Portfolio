## Descripcion
En este punto se hace una descripcion de las variables que componen el set de datos AB_NYC_2019.csv

El set de datos AB_NYC_2019.csv se compone de 48.895 regitros con 16 variables(columnas) cada columna tiene las siguientes caracterisiticas:

- id: variable cualitativa nominal
- name: variable cualitativa nominal
- host_id: variable cualitativa nominal
- host_name: variable cualitativa nominal
- neighbourhood_group: cualitativa nominal
- neighbourhood: cualitativa nominal
- latitude: variable cuantitativa continua
- longitude: variable cuantitativa continua
- room_type: variable cualitativa ordinal
- price: variable cuantitativa continua - Rango ($ 0 a $ 10.000)
- minimum_nights: variable cuantitativa discreta - Rango (1 a 1.250)
- number_of_reviews: variable cuantitativa discreta - Rango (0 a 629)
- last_review: variable cualitativa ordinal
- reviews_per_month: variable cuantitativa continua - Rango (0.01 a 58.5)
- calculated_host_listings_count: variable cuantitativa discreta - Rango (1 a 327)
- availability_365: variable cuantitativa discreta - Rango (0 a 365)
