### Summary  

The Iris dataset consists of 150 observations. In this dataset, there are 3 types of flowers identified in the variable *Species*. For each observation, 4 measurements were taken: petal length and width, and sepal length and width. The goal of this dataset is to relate the flower measurements to its species. Below is a brief summary:

|Variable|Description|Type|  
|---|---|---|  
|Sepal.Length|Sepal length (cm)|Quantitative continuous|  
|Sepal.Width|Sepal width (cm)|Quantitative continuous|  
|Petal.Length|Petal length (cm)|Quantitative continuous|  
|Petal.Width|Petal width (cm)|Quantitative continuous|  
|Species|Flower species|Qualitative nominal|  

```python
# We'll import all nedeed libraries
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Import data
variables = ['sepal_length','sepal_width','petal_length','petal_width','species']
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names= variables) # dataset
```

## Matplotlib
This is a standart library from python, mainly used to create visualizations, we can create basic visualizations as well as complex ones.

```python
plt.hist(iris['sepal_length'])

plt.xlabel('Intervalo fecuencia de longitud de sepalo')
plt.ylabel('Conteo de frecuencias')
plt.title('Grafico de frecuencias')
plt.show()
```
