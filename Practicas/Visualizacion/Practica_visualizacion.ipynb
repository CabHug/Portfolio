### Summary  

The Iris dataset consists of 150 observations. In this dataset, there are 3 types of flowers identified in the variable *Species*. For each observation, 4 measurements were taken: petal length and width, and sepal length and width. The goal of this dataset is to relate the flower measurements to its species. Below is a brief summary:

|Variable|Description|Type|  
|---|---|---|  
|Sepal.Length|Sepal length (cm)|Quantitative continuous|  
|Sepal.Width|Sepal width (cm)|Quantitative continuous|  
|Petal.Length|Petal length (cm)|Quantitative continuous|  
|Petal.Width|Petal width (cm)|Quantitative continuous|  
|Species|Flower species|Qualitative nominal|  

```python
# We'll import all nedeed libraries
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Import data
variables = ['sepal_length','sepal_width','petal_length','petal_width','species']
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names= variables) # dataset
```

## Matplotlib
This is a standart library from python, mainly used to create visualizations, we can create basic visualizations as well as complex ones.

```python
plt.hist(iris['sepal_length'])

plt.xlabel('Intervalo fecuencia de longitud de sepalo')
plt.ylabel('Conteo de frecuencias')
plt.title('Grafico de frecuencias')
plt.show()
```

```python
plt.plot(iris['sepal_length'])
```

```python
plt.boxplot(iris['sepal_length'])
```

```python
plt.bar(iris['species'], iris['sepal_length'])
```

```python
plt.barh(iris['species'], iris['sepal_length'])
```

Now we'll use the object method in matplotlib speacially useful if we whant to show different graps
on a canvas. this logic involves create an object "figure" after we can use some methos to modify it's behabior

```python
fig, axes = plt.subplots()

axes.hist(iris['sepal_length'], 
          alpha = 0.7)
axes.set_xlabel('intervalo de frecuencia')
axes.set_ylabel('frecuencia acumulada')
```

```python
# Now we will create a new canvas wit different layout

#fig, axis = plt.subplots((filas),(columnas), figsize = (6,6))
fig, axes = plt.subplots(1,2, figsize = (8,6))

# we can configure each of the axis individually
axes[0].hist(iris['sepal_length'])
axes[0].set_xlabel('Intervalo de freciencia')
axes[0].set_ylabel('Frecuencia acumulada')

axes[1].boxplot(iris['sepal_length'])
axes[1].set_xlabel('Intervalo de freciencia')
axes[1].set_ylabel('Frecuencia acumulada')
```

```python
iris.groupby('species').mean() # Crear una tabla agrupando por la especie de flor, para cada combinacin de variables calcula la media
```

```python
iris.groupby('species').mean().plot(kind='bar')

plt.xlabel('Especies')
plt.ylabel('Promedio de mediciones')
```

```python
fig, axes = plt.subplots(1,2, figsize = (10,4))

axes[0].hist(iris['sepal_length'],
    bins = 15,
    color = 'violet');
axes[0].set_xlabel('Intervalo de freciencia')
axes[0].set_ylabel('Frecuencia acumulada')

axes[1].scatter(iris['sepal_length'],
    iris['petal_length'],
    marker = 'x',
    color = 'red');
axes[1].set_xlabel('Intervalo de freciencia')
axes[1].set_ylabel('Frecuencia acumulada')
```

### Now I'll use the .groupby() method from pandas to create a new data frame and also generate new plots from this

```python
iris.groupby('species').mean()
```

```python
iris.groupby('species').mean()['petal_length'].plot(kind='bar')
plt.xlabel('Ancho del spalo (cm)')
plt.ylabel('Especie de la flor')
```
