### Summary  

The Iris dataset consists of 150 observations. In this dataset, there are 3 types of flowers identified in the variable *Species*. For each observation, 4 measurements were taken: petal length and width, and sepal length and width. The goal of this dataset is to relate the flower measurements to its species. Below is a brief summary:

|Variable|Description|Type|  
|---|---|---|  
|Sepal.Length|Sepal length (cm)|Quantitative continuous|  
|Sepal.Width|Sepal width (cm)|Quantitative continuous|  
|Petal.Length|Petal length (cm)|Quantitative continuous|  
|Petal.Width|Petal width (cm)|Quantitative continuous|  
|Species|Flower species|Qualitative nominal|  

```python
# We'll import all nedeed libraries
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Import data
variables = ['sepal_length','sepal_width','petal_length','petal_width','species']
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names= variables) # dataset
```

## Matplotlib
This is a standart library from python, mainly used to create visualizations, we can create basic visualizations as well as complex ones.

```python
plt.hist(iris['sepal_length'])

plt.xlabel('Intervalo fecuencia de longitud de sepalo')
plt.ylabel('Conteo de frecuencias')
plt.title('Grafico de frecuencias')
plt.show()
```

Now we'll use the object method in matplotlib speacially useful if we whant to show different graps
on a canvas. this logic involves create an object "figure" after we can use some methos to modify it's behabior

```python
fig, axes = plt.subplots()

axes.hist(iris['sepal_length'], 
          alpha = 0.7)
axes.set_xlabel('intervalo de frecuencia')
axes.set_ylabel('frecuencia acumulada')
```

```python
# Now we will create a new canvas wit different layout

#fig, axis = plt.subplots((filas),(columnas), figsize = (6,6))
fig, axes = plt.subplots(1,2, figsize = (8,6))

# we can configure each of the axis individually
axes[0].hist(iris['sepal_length'])
axes[0].set_xlabel('Intervalo de freciencia')
axes[0].set_ylabel('Frecuencia acumulada')

axes[1].boxplot(iris['sepal_length'])
axes[1].set_xlabel('Intervalo de freciencia')
axes[1].set_ylabel('Frecuencia acumulada')
```

```python
iris.groupby('species').mean() # Crear una tabla agrupando por la especie de flor, para cada combinacin de variables calcula la media
```

```python
iris.groupby('species').mean().plot(kind='bar')

plt.xlabel('Especies')
plt.ylabel('Promedio de mediciones')
```

```python
fig, axes = plt.subplots(1,2, figsize = (10,4))

axes[0].hist(iris['sepal_length'],
    bins = 15,
    color = 'violet');
axes[0].set_xlabel('Intervalo de freciencia')
axes[0].set_ylabel('Frecuencia acumulada')

axes[1].scatter(iris['sepal_length'],
    iris['petal_length'],
    marker = 'x',
    color = 'red');
axes[1].set_xlabel('Intervalo de freciencia')
axes[1].set_ylabel('Frecuencia acumulada')
```

### Now I'll use the .groupby() method from pandas to create a new data frame and also generate new plots from this

```python
iris.groupby('species').mean()['petal_length'].plot(kind='bar')
plt.xlabel('Ancho del spalo (cm)')
plt.ylabel('Especie de la flor')
```

# SEABORN

Seaborn is a useful library from python that allows me to develop great visualizations, this is build over matplotlib therefore we can use some similar features from this on seaborn

```python
# We define a default stile
sns.set_theme('notebook', palette = 'bright')
```

## Histogram
This is a useful tool to know cuantitative data, allow us to perform univariate and bivariate analysis, also, allow us know the data distribution

```python
sns.histplot(
    data= iris,
    x = 'sepal_length',
    bins= 15,
    kde = True
)
```

```python
# Coding a feature inside our plot

sns.histplot(
    data= iris,
    x = 'sepal_length',
    bins = 10,
    kde = True,
    hue = 'species'
)
```

```python
# now I will create a plot with multiple axis

fig, axis = plt.subplots(2, 1, figsize =(6,10))

sns.histplot(
    data= iris,
    x = 'sepal_length',
    kde = True,
    ax= axis[0]
)

axis[0].set_xlabel('Largo del spalo (cm)')
axis[0].set_ylabel('Frecuencia')

sns.scatterplot(
    data= iris,
    x= 'sepal_length',
    y = 'petal_length',
    hue = 'species',
    ax = axis[1]
)
axis[1].set_xlabel('Largo del spalo (cm)')
axis[1].set_ylabel('Largo del petalo (cm)')
```

```python
# Kernel densiti estimation plot

sns.kdeplot(
    data= iris,
    x = 'petal_length',
    hue= 'species',
    multiple= 'stack'
)
```

### Bar Diagram

A bar chart is useful to know the categorical variables, it allow us cout those variables, also we can relate to another cuantitative variable

```python
sns.countplot(
    data= iris,
    x = 'species'
)
```

```python
sns.boxplot(
    data = iris,
    y = 'petal_length',
    x = 'species',
    hue = 'species'
)
```

```python
fig, axi = plt.subplots(1, 2, figsize=(10,8))

sns.boxplot(
    data= iris,
    x= 'species',
    y= 'petal_length',
    hue= 'species',
    ax= axi[0]
)
axi[0].set_xlabel('species')
axi[0].set_ylabel('petal length (cm)')

sns.boxplot(
    data= iris,
    x= 'sepal_length',
    y= 'species',
    hue= 'species',
    ax= axi[1]
)
axi[1].set_xlabel('species')
axi[1].set_ylabel('sepal length (cm)')
```

### Scatterplot
A scatterplot allows me compare two categorical variables, then I can see it's relation

```python
sns.scatterplot(
    data= iris,
    x = 'petal_length',
    y = 'sepal_length',
    hue= 'species'
)
```

```python
# A really useful method is lmplot(), this one allows me perform linear regresion models

sns.lmplot(
    data= iris,
    x= 'petal_length',
    y= 'sepal_length'
)
```

### Pairplot

Another very useful chart, it allows me show the relation between all variables inside my dataframe

```python
sns.pairplot(
    data= iris
)
```
